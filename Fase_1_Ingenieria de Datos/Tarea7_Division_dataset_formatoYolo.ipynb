```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Ruta de tu dataset YA ORGANIZADO
RUTA_DATASET = r"C:\Users\GARI\Documents\DATASETS\DATASET_ORINAL_DESCARTADAS\arboles_dataset_yolo"

paths = {
    "train": os.path.join(RUTA_DATASET, "images", "train"),
    "val": os.path.join(RUTA_DATASET, "images", "val"),
    "test": os.path.join(RUTA_DATASET, "images", "test"),
}

EXT = (".jpg", ".jpeg", ".png")

# Contar im치genes por conjunto
conteos = {}
for subset, ruta in paths.items():
    conteos[subset] = len([f for f in os.listdir(ruta) if f.lower().endswith(EXT)])

total_imagenes = sum(conteos.values())

# Crear tabla
df_particion = pd.DataFrame({
    "Conjunto": ["Entrenamiento (70%)", "Validaci칩n (20%)", "Prueba (10%)"],
    "Cantidad": [conteos["train"], conteos["val"], conteos["test"]],
    "Porcentaje real": [
        round(conteos["train"] / total_imagenes * 100, 2),
        round(conteos["val"] / total_imagenes * 100, 2),
        round(conteos["test"] / total_imagenes * 100, 2),
    ]
})

df_particion
```

```python

labels = [
    f"Train: {conteos['train']} im치genes ({df_particion['Porcentaje real'][0]}%)",
    f"Val: {conteos['val']} im치genes ({df_particion['Porcentaje real'][1]}%)",
    f"Test: {conteos['test']} im치genes ({df_particion['Porcentaje real'][2]}%)"
]

sizes = [conteos["train"], conteos["val"], conteos["test"]]

plt.figure(figsize=(7,7))
plt.pie(
    sizes,
    labels=labels,
    autopct="%1.1f%%",
    startangle=90,
    textprops={'fontsize': 10}
)
plt.title("Distribuci칩n del conjunto de datos etiquetado (70%  20%  10%)")
plt.show()
```
