```python
import gradio as gr
from ultralytics import YOLO
from PIL import Image
```

```python
# 2) Cargar TU modelo entrenado
# Cambia la ruta si tu best.pt está en otra carpeta
model = YOLO(r'C:\Users\GARI\Documents\IPYNB\PRUEBAS_MODELO_YOLOV8\07-prueba_data_1024_yolov8L_ADAMS_LR_FINE_T_VF\yolov8l_1024_precision_finetune_finalAdamW2\weights\best.pt')


def detectar_arboles(image, conf_thres):
    """
    image: PIL.Image (la imagen que sube el usuario)
    conf_thres: umbral de confianza para las detecciones
    """
    # 3) Ejecutar el modelo
    results = model.predict(image, conf=conf_thres, verbose=False)
    res = results[0]

    # 4) Imagen anotada (con bounding boxes)
    # res.plot() devuelve un ndarray en BGR
    annotated = res.plot()
    annotated = annotated[:, :, ::-1]  # BGR -> RGB

    # 5) Conteo de árboles (cantidad de bounding boxes)
    tree_count = len(res.boxes)

    texto = f"Árboles detectados: {tree_count}"
    return Image.fromarray(annotated), texto
```

```python

# 6) Interfaz con Gradio (lo más sencillo para empezar)
demo = gr.Interface(
    fn=detectar_arboles,
    inputs=[
        gr.Image(type="pil", label="Sube una imagen aérea"),
        gr.Slider(0.1, 0.9, value=0.5, step=0.05, label="Confianza mínima"),
    ],
    outputs=[
        gr.Image(type="pil", label="Imagen con detecciones"),
        gr.Textbox(label="Conteo de árboles"),
    ],
    title="Detector de árboles YOLOv8",
    description="Sube una imagen urbana y el modelo marcará las copas detectadas y mostrará el conteo."
)

if __name__ == "__main__":
    demo.launch()
```
