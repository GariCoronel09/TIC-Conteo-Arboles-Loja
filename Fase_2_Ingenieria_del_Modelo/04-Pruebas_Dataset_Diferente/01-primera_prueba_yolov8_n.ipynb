```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt

# Para que las imágenes salgan dentro del notebook (normalmente ya es así)
%matplotlib inline
```

```python
# Rutas
model = YOLO(r"C:\Users\GARI\Documents\IPYNB\PRUEBAS_MODELO_YOLOV8\01-modelo1_yolov8n_prueba_experimental\yolov8n_conteo_arboles2\weights\best.pt")
#img_path = r"C:\Users\GARI\Documents\DATASETS\m1-only-src\images\DJI_0438.jpg"
img_path = r"C:\Users\GARI\Documents\DATASETS\m1-only-src\images\DJI_0343.jpg" # DJI_0343
# Inferencia (sin imprimir texto extra)
results = model(img_path, verbose=False)
```

```python
# Conteo
num_arboles = len(results[0].boxes)
print("Yolov8n-Árboles detectados:", num_arboles)

# Imagen anotada con las cajas de YOLO
annotated = results[0].plot()   # OJO: sin show=True

# Mostrar dentro del notebook
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB))
plt.axis("off")
```

```python
# AQUÍ CAMBIAS EL TÍTULO
plt.title(f"Yolov8n - Número de árboles detectados: {num_arboles}", fontsize=18)
plt.show()
```
